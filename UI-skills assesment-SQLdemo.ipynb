```python

#Dup Primary ID is an error
names1=names.drop_duplicates()
names1.describe(include='all')
```

```python
names2=names1.drop_duplicates(subset=['HouseholdID'])
names2.describe(include='all')
```

## Create merged file for exploritory market sementation in Power BI

```python

donors=pd.merge(contribution, people, on=['PrimaryID'], how='outer', indicator=True)
donors=pd.merge(donors, names2, on=['PrimaryID'], how='outer',)
donors=donors.sort_values(by=['Amount','State','College','Department',],ascending=False)
donors.to_excel('part2BI.xlsx', index=False)
print(donors.isnull().sum())
#donors.describe(include='all')
#donors.head(10)
```

```python
donors.to_sql("DONORTBL", con, if_exists='replace', index=False,)#method="multi")
donors.head()
```

```python
%sql create table DONORTABLE as select * from DONORTBL where Date is not null
```

```python
topsingle=%sql SELECT firstname, lastname, PartnerStatus, State, PersonStatus, Affiliation, College, Department,  Purpose, SUM(recognitioncredit) AS totAmount, COUNT(date) AS freq FROM DONORTABLE WHERE RecognitionCredit >= 1000000 GROUP BY PrimaryID ORDER BY recognitioncredit DESC
topsingle=pd.DataFrame(topsingle)
topsingle.to_excel('SQLtopsingleDonors.xlsx')
```

```python
top=%sql SELECT firstname, lastname, PartnerStatus, State, PersonStatus, Affiliation, College, Department,  Purpose, SUM(recognitioncredit) AS totAmount, COUNT(date) AS freq FROM DONORTABLE GROUP BY PrimaryID ORDER BY totAmount DESC
top=pd.DataFrame(top)
top.to_excel('SQLtopOverallDonors.xlsx')
```

```python
loyal=%sql SELECT firstname, lastname, PartnerStatus, state, PersonStatus, Affiliation, College, Department,  Purpose, SUM(recognitioncredit) AS totAmount, COUNT(date) AS freq FROM DONORTABLE GROUP BY PrimaryID ORDER BY freq DESC LIMIT 250
loyal=pd.DataFrame(loyal)
loyal.to_excel('SQLLoyaldonors.xlsx')
loyal
```
